```python
import cv2    
import matplotlib.pyplot as plt
#import seaborn as sns
```

# 1
### read an image

```python
#plt.style.use('seaborn')

img = cv2.imread('IMG_20201226_114356.jpg')
plt.imshow(img)
plt.show()
# reads the (R G B) as (B G R) 
```

# 2

```python
# convert between two color 
real_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(real_img)
plt.show()
```

```python
# if all are 000000 this means color is black 
# if all are 255 255 255 the color is white
print(real_img)
```

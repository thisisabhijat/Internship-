# Distribution plot

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

# Let's get started with distribution plot

```python
plt.style.use('dark_background')

sns.distplot(tips['total_bill'], color = 'yellow')
plt.show()
```

```python
# distplot is soon going to be removed as is asked to use displot in it's place
sns.displot(tips['total_bill'], color = 'red')
plt.show()
```

```python
# for removing the line of distplot by using the function kde
sns.distplot(tips['total_bill'], color = 'white', kde = False)
plt.show()
```

```python
# increasing the size using the bin function
sns.distplot(tips['total_bill'], color = 'blue', kde = False, bins = 40)
plt.show()
```

# KDE Plot

```python
plt.style.use('dark_background')

sns.kdeplot(tips['total_bill'], color = 'orange')
plt.show()
```

# Joint Plot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, color = 'green')
plt.show()
```

```python
# changing the orientation the above figure has scattered by default, done by using the kind function
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, color = 'pink', kind = 'hex')
plt.show()
```

```python
plt.style.use('dark_background')

sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, color = 'brown', kind = 'reg')
plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, color = 'gold', kind = 'scatter')
plt.show()
```

# Pair Plot

```python
plt.style.use('dark_background')

sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue = 'sex')
plt.show()
```

```python
# matplotlib for creating static, animated, and interactive visualizations in Python
# pyplot is a collection of functions that make matplotlib work

import matplotlib.pyplot as plt
import numpy as np
```

```python
# line plotting in graph
x = np.arange(10) # x is an array of 10 numbers 
y1 = x ** 2 # y is squares of all the values of x
y2 = 2*x + 3 # equation of a line 

print(x)
print(y1)
print(y2)
```

```python
# themes which the matplotlib provides
themes = plt.style.available
print(themes)
```

```python
# separate graphs
plt.style.use('seaborn')
plt.style.use('dark_background')

plt.plot(x)
plt.show()

plt.plot(y1)
plt.show()

plt.plot(y2)
plt.show()
```

```python
# applying themes on my graph

# combined plotting of graphs using plot() function and i will add a new color to the line by adding a new variable next to (x, y1,....)
plt.plot(x, y1, color = 'red')
plt.show()

plt.plot(x, y2, color = 'gold')
plt.show()
```

```python
# now a combined graph
plt.plot(x, y1, color = 'red')
plt.plot(x, y2, color = 'gold')

# plt.plot
plt.show()
```

```python
# giving titles to the axes using the xlabel() and ylabel() commands 
plt.plot(x, y1, color = 'red')
plt.plot(x, y2, color = 'gold')

plt.xlabel("x = np.arange(10)", color = 'pink')
plt.ylabel("y1 = x ** 2 &&& y2 = 2*x + 3", color = 'orange')

plt.show()
```

```python
# denoting which graph is for what by using label like APPLE AND GRAPES and using legend funcion
plt.plot(x, y1, color = 'red', label = "APPLE")
plt.plot(x, y2, color = 'gold', label = "GRAPES")

plt.xlabel("x = np.arange(10)", color = 'pink')
plt.ylabel("y1 = x ** 2 &&& y2 = 2*x + 3", color = 'orange')
plt.title("Mathematical Graphs", color = 'white')
plt.legend()

plt.show()
```

```python
# now making GRAPES a dashed line by using the function linestyle and marker 
plt.plot(x, y1, color = 'red', label = "APPLE", linestyle = 'dashed', marker = 'o')
plt.plot(x, y2, color = 'gold', label = "GRAPES", linestyle = 'dashed', marker = '*')

plt.xlabel("x = np.arange(10)", color = 'pink')
plt.ylabel("y1 = x ** 2 &&& y2 = 2*x + 3", color = 'orange')
plt.title("Mathematical Graphs", color = 'white')
plt.legend()

plt.show()
```

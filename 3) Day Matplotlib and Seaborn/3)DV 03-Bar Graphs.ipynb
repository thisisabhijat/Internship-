```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creating the bar graph
plt.style.use("dark_background")

plt.bar([0, 1, 2], [10, 20, 15], color = 'gold')

plt.show()
```

```python
# comparing the prices of two years
plt.bar([0, 1, 2], [10, 20, 15], color = 'gold')# current year
plt.bar([0, 1, 2], [20, 10, 12], color = 'silver')# next year

plt.show()
```

```python
# removing the overlapping
x_coordinates = np.array([0, 1, 2])

plt.bar(x_coordinates, [10, 20, 15], color = 'gold')# current year
plt.bar(x_coordinates + 0.5, [20, 10, 12], color = 'silver')# next year

plt.show()
```

```python
# removing further overlapping 
x_coordinates = np.array([0, 1, 2])

plt.bar(x_coordinates, [10, 20, 15], color = 'gold', width = 0.5)# current year
plt.bar(x_coordinates + 0.5, [20, 10, 12], color = 'silver', width = 0.5)# next year

plt.show()
```

```python
# further making it more clear by multiplying it by 2 and adding label
x_coordinates = np.array([0, 1, 2]) * 2

plt.bar(x_coordinates, [10, 20, 15], color = 'gold', width = 0.5, label = "current year")# current year
plt.bar(x_coordinates + 0.5, [20, 10, 12], color = 'silver', width = 0.5, label = "next year")# next year

plt.legend()
plt.show()
```

```python
# giving trick_label
x_coordinates = np.array([0, 1, 2]) * 2

plt.bar(x_coordinates, [10, 20, 15], color = 'gold', width = 0.5, label = "current year", tick_label = ["Gold", "Platinum", "Silver"])# current year
plt.bar(x_coordinates + 0.5, [20, 10, 12], color = 'silver', width = 0.5, label = "next year")# next year

plt.legend()
plt.show()
```

```python
# titles to coordinates
plt.style.use("dark_background")
x_coordinates = np.array([0, 1, 2]) * 2

plt.bar(x_coordinates, [10, 20, 15], color = 'gold', width = 0.5, label = "current year", tick_label = ["Gold", "Platinum", "Silver"])# current year
plt.bar(x_coordinates + 0.5, [20, 10, 12], color = 'red', width = 0.5, label = "next year")# next year
plt.title("comparing prices of costly elements")
plt.xlabel("Elements")
plt.ylabel("Price")
plt.ylim(0, 30)

plt.legend()
plt.show()
```

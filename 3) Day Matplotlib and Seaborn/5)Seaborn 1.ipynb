```python
import seaborn as sns
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# entire data frame using tips used for visualization
tips = sns.load_dataset('tips')
tips
```

```python
# head is used to show the first 5 data frames
tips.head()
```

# Bar plot 
along with that we will study aggreation funcitons in bar plot like estimator = np.std, estimator = np.var

```python
# creating bar plot with mean in y axis
plt.style.use('seaborn')
plt.style.use('dark_background')

sns.barplot(x = 'sex', y = 'total_bill', data = tips, alpha = 0.5)# alpha is used for opacity level

plt.show()
```

```python
# adding estimator will show the standard deviation in y axis
plt.style.use('dark_background')

sns.barplot(x = 'sex', y = 'total_bill', data = tips, color = 'red', estimator = np.std)

plt.show()
```

```python
# estimator shows the variance in y axis
plt.style.use('dark_background')

sns.barplot(x = 'sex', y = 'total_bill', data = tips, color = 'red', estimator = np.var)

plt.show()
```

# Count Plot

```python
# creating the countplot here you cant change the mean to standard deviation in count thus it is not required to give the ylabel
sns.countplot(x = 'sex', data = tips, color = 'yellow')

plt.show()
```

```python
sns.countplot(x = 'time', data = tips, color = 'pink')
plt.show()
```

# Box Plot

```python
# box plot with hue function
plt.style.use('dark_background')

sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker', color = 'red')
plt.show()
```

```python
# box plot
plt.style.use('seaborn')

sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', color = 'gold')
plt.show()
```

# Violin Plot

```python
plt.style.use('dark_background')

sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True, color = 'green')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# loading the dataset tips
tips = sns.load_dataset('tips')
tips
```

```python
tips.head()
```

```python
# Correlation function shows relation between numeric values
tips_corr = tips.corr()
print(tips_corr)
```

```python
tips.corr()
```

# Heat Map

```python
plt.style.use('dark_background')

sns.heatmap(tips_corr)
plt.show()
```

```python
# displaying the numbers with parameter annot
sns.heatmap(tips_corr, annot = True)
plt.show()
```

```python
# loading the dataset flights just like tips
flights = sns.load_dataset('flights')
flights
```

```python
print(flights)
```

```python
flights_pivot = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
flights_pivot
```

```python
sns.heatmap(flights_pivot)
plt.show()
```

```python
# colormap usage
sns.heatmap(flights_pivot, cmap = 'coolwarm')
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# simple pie chart 
plt.style.use("dark_background")
subjects = "Maths", "Chemistry", "Physics", "English"
weightage = [20, 10, 15, 5]

plt.pie(weightage, labels = subjects)
plt.show()
```

```python
# the exploding ie removing a fixed part from the chart
plt.style.use("dark_background")

plt.pie(weightage, labels = subjects, explode = (0, 0 ,1, 0))
plt.show()
```

```python
# decreasing the exploding part
plt.style.use("dark_background")

plt.pie(weightage, labels = subjects, explode = (0, 0.5, 0.1, 0))
plt.show()
```

```python
# with % symbols
plt.pie(weightage, labels = subjects, autopct = '%1.1f%%')# autopercentage
plt.show()
```

```python
# now showing the shadow
plt.style.use('seaborn')

plt.pie(weightage, labels = subjects, autopct = '%1.1f%%', shadow = True)
plt.title("Subjects")
plt.show()
```

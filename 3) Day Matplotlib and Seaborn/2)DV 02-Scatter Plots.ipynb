```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# for the background color
plt.style.use('dark_background')
```

```python
# made an array 
prices = np.array([1, 2, 3, 4]) ** 3
print(prices)
```

```python
# the graph is becoming in y axis by default and will consider x as cordinates like y = f(x)
plt.plot(prices, linestyle = 'dashed', marker = 'o')
plt.show()
```

# Scatters Plots

```python
plt.style.use('dark_background')
```

```python
x = np.arange(10) # x is an array of 10 numbers 
y1 = x ** 2 # y is squares of all the values of x
y2 = 2*x + 3 # equation of a line 
```

```python
# difference between value and graph
print(x)

plt.plot(x)
plt.show()
```

```python
# finally using scatter function
plt.scatter(x, y1, color = 'gold')
plt.scatter(x, y2, color = 'red')
plt.show()
```

# Adjusting the size of the plot

```python
# this figure function is used to decide the figure
plt.figure(figsize = (5, 5))

plt.plot(x, y1, color = 'red', label = "APPLE", linestyle = 'dashed', marker = 'o')
plt.plot(x, y2, color = 'gold', label = "GRAPES", linestyle = 'dashed', marker = '*')

plt.xlabel("x = np.arange(10)", color = 'pink')
plt.ylabel("y1 = x ** 2 &&& y2 = 2*x + 3", color = 'orange')
plt.title("Mathematical Graphs", color = 'white')
plt.legend()

plt.show()
```

```python
# changing the plot with scatter
plt.figure(figsize = (5, 5))# figures can be changed

plt.scatter(x, y1, color = 'red', label = "APPLE", linestyle = 'dashed', marker = 'o')
plt.scatter(x, y2, color = 'gold', label = "GRAPES", linestyle = 'dashed', marker = '*')

plt.xlabel("x = np.arange(10)", color = 'pink')
plt.ylabel("y1 = x ** 2 &&& y2 = 2*x + 3", color = 'orange')
plt.title("Mathematical Graphs", color = 'white')
plt.legend()

plt.show()
```

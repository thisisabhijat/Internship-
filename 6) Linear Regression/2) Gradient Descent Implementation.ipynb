```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2

print(X, Y)
```

```python
plt.style.use('seaborn')

plt.plot(X, Y)
plt.xlabel("X", color = 'red')
plt.ylabel("F(X)", color = 'red')
plt.show()
```

 # 1) Given a function f(x), we want to find the value of x that minimizes f
- Using Gradient Descent Algorithm


```python
x = 0
# step size or learning grade
lr = 0.1

# take 50 steps in downhill direction
for i in range(50):
    
    # gradient is differentiation of (x-5)**2 is 2*(x-5) 
    grad = 2*(x-5)
    
    x = x - lr * grad
    print(x)
```

# 2 Adding more visualization to our plot

```python
plt.style.use('dark_background')

x = 0
lr = 0.1
# fx denotes error and our goal is to minimize error
error = []

plt.plot(X, Y, color = 'Yellow')

for i in range(50):
    grad = 2*(x-5)
    x = x - lr*grad
    
    y_or_e = (x-5)**2
    error.append(y_or_e)
    
    plt.scatter(x, y_or_e)
    print(x)
```

```python
# step size is continuously decresing because ie distance between the points is decreasing 
# the reason is d(f(x))/dx or dy/dx is continuously is decreasing
```

```python
plt.plot(error, color = 'red')
plt.show()
```

# Surface plots | Data Visualization
## Surface plots are used to 
- Visualize loss function in Machine Learning & Deep Learning 
- Visualize state or state value functions in Reinforcement Learning

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1)

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6, 7])

a,b = np.meshgrid(a, b)
print(a)
print(b)
```

# 2)-Surface plot(plot_surface)

```python
plt.style.use('dark_background')
```

```python
# generating surface plot
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a, b, a**2+b**2, cmap = 'rainbow')
# red color denotes the region where the value is high 
# blue color denotes where the value is low
plt.show()
```

### practise example

```python
a = np.arange(-1, 1, 0.02)
b = a

a, b = np.meshgrid(a, b)

fig = plt.figure(figsize = (10, 10))
axes = fig.gca(projection = '3d')
axes.plot_surface(a, b, a**2+b**2, cmap = 'rainbow')
plt.title('Surface Plot')
plt.show()
```

# 3)-Contours

```python
fig = plt.figure(figsize = (10, 10))
axes = fig.gca(projection = '3d')
axes.contour(a, b, a**2+b**2, cmap = 'rainbow')
plt.title('Contour Plot')
plt.show()
```

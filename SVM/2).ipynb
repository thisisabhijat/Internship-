```python
from sklearn.datasets import make_circles
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
X,Y = make_circles(n_samples = 500, noise = 0.05)
```

```python
print(X.shape, Y.shape)
```

```python
plt.grid(True)
plt.scatter(X[:, 0], X[:, 1], c = Y)
plt.show()
```

```python
def phi(X):
    #Non Linear Tranformation 
    
    X1 = X[:,0]
    X2 = X[:, 1]
    X3 = X1**2 + X2**2
    
    X_ = np.zeros((X.shape[0], 3))
    print(X_.shape)
    
    X_[:,:-1] = X
    X_[:,-1] = X3
    
    return X_
```

```python
X_ = phi(X)
```

```python
print(X[:3, :])
```

```python
print(X_[:3,:])
```

```python
def plot3d(X, show = True):
    fig = plt.figure(figsize = (10, 10))
    ax = fig.add_subplot(111, projection = '3d')
    X1 = X[:,0]
    X2 = X[:,1]
    X3 = X[:,2]
    
    ax.scatter(X1, X2, X3, zdir = 'z', s = 20, c = Y, depthshade = True)
    
    if (show == True):
        plt.show()
    return ax
```

```python
ax = plot3d(X_)
plt.show()
```

# Logistic Classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
lr = LogisticRegression()
```

```python
acc = cross_val_score(lr, X, Y, cv = 5).mean()
print("Accuracy X(2D) is %.4f"%(acc*100))
```

# Logistic Classfier on Higher Dimension Space

```python
acc = cross_val_score(lr, X_, Y, cv = 5).mean()
print("Accuracy X(3D) is %.4f"%(acc*100))
```

### visualise decision 

```python
lr.fit(X_, Y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
wts = lr.coef_
print(wts)
```

```python
bias = lr.intercept_
print(bias)
```

```python
xx, yy = np.meshgrid(range(-2,2), range(-2,2))
print(xx)
print(yy)
```

```python
z = -(wts[0,0]*xx + wts[0,1]**yy+bias)/wts[0,2]
print(z)
```

```python
ax = plot3d(X_, False)
ax.plot_surface(xx, yy, z)
plt.show()
```

# The correct one 

```python
X,Y = make_circles(n_samples=500,noise=0.02)
```

```python
print(X.shape,Y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
def phi(X):
    """"Non Linear Transformation"""
    X1 = X[:,0]
    X2 = X[:,1]
    X3 = X1**2 + X2**2
    
    X_ = np.zeros((X.shape[0],3))
    print(X_.shape)
    
    X_[:,:-1] = X
    X_[:,-1] = X3

    return X_
```

```python
X_ = phi(X)
```

```python
print(X[:3,:])
```

```python
print(X_[:3,:])
```

```python
def plot3d(X,show=True):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111,projection='3d')
    X1 = X[:,0]
    X2 = X[:,1]
    X3 = X[:,2]
    
    ax.scatter(X1,X2,X3,zdir='z',s=20,c=Y,depthshade=True)
    
    if(show==True):
        plt.show()
    return ax
```

```python
ax = plot3d(X_)
```

```python
lr = LogisticRegression()
```

```python
acc = cross_val_score(lr,X,Y,cv=5).mean()
print("Accuracy X(2D) is %.4f"%(acc*100))
```

```python
acc = cross_val_score(lr,X_,Y,cv=5).mean()
print("Accuracy X(3D) is %.4f"%(acc*100))
```

```python
lr.fit(X_,Y)
```

```python

wts = lr.coef_
print(wts)
```

```python

bias = lr.intercept_
```

```python

xx,yy = np.meshgrid(range(-2,2),range(-2,2))
print(xx)
print(yy)
```

```python
z = -(wts[0,0]*xx + wts[0,1]*yy+bias)/wts[0,2]
print(z)
```

```python
ax = plot3d(X_,False)
ax.plot_surface(xx,yy,z,alpha=0.2)
plt.show()
```

```python
# custom kernel
```

```python
def custom_kernel(x1, x2):
    return np.square(np.dot(x1, x2.T))

svc = svm.SVC(kernel = custom_kernel)
svc.fit(X,Y)
svc.score(X,Y) * 100
```

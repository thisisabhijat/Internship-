# Pandas 
- Task = Make a scatter plot b/w length and movie title

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# 1

```python
df = pd.read_csv('movie_metadata.csv')
df.head(3)
```

# 2) 

```python
# columns converted into array
columns_array = np.array(df.columns)
print(columns_array)
```

# 3

```python
# getting the titles
titles = list(df.get('movie_title'))
print(titles)
```

# 4

```python
# using if else
freq_titles = {}

for title in titles:
    length = len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
        
print(freq_titles)
```

```python
# using try and except 
freq_titles = {}

for title in titles:
    length = len(title)
    
    try:
        freq_titles[length] += 1
    except:
        freq_titles[length] = 1
print(freq_titles)
```

# 5

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
```

```python
# creating the plot
plt.style.use('seaborn')
plt.style.use('dark_background')

plt.scatter(X, Y, color = 'yellow')
plt.xlabel('length')
plt.ylabel('num_of_movies')
plt.title('length to num_of_movies')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
# the yellow color shows that x is not related to y, since the sigma(xy) = 0
plt.style.use('seaborn')
plt.style.use('dark_background')

mean = np.array([0.0, 0.0])
cov = np.array([[1, 0],
                [0, 1]])
mean2 = np.array([5.0, 6.0])
cov2 = np.array([[1, 0.5],
                 [0.5, 1]])
mean3 = np.array([10.0, 12.6])
cov3 = np.array([[1, 0.8], 
                [0.8, 1]])

dist = np.random.multivariate_normal(mean, cov, 500)
dist2 = np.random.multivariate_normal(mean2, cov2, 500)
dist3 = np.random.multivariate_normal(mean3, cov3, 500)

plt.scatter(dist[:, 0], dist[:, 1], color = 'yellow')
plt.scatter(dist2[:, 0], dist2[:, 1], color = 'red')
plt.scatter(dist3[:, 0], dist3[:, 1], color = 'skyblue')

plt.show()
```

```python
plt.style.use('seaborn')
plt.style.use('dark_background')

mean = np.array([0.0, 0.0])
cov = np.array([[1, 0.8],
                [0.8, 1]])
mean2 = np.array([5.0, 6.0])
cov2 = np.array([[1.3, 0.3],
                 [0.3, 1.1]])


dist = np.random.multivariate_normal(mean, cov, 500)
dist2 = np.random.multivariate_normal(mean2, cov2, 500)

plt.scatter(dist[:, 0], dist[:, 1], color = 'yellow')
plt.scatter(dist2[:, 0], dist2[:, 1], color = 'red')

plt.show()
```
